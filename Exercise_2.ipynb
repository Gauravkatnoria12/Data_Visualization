```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
```

```python
df = pd.DataFrame(data)
```

```python
df
```

#  **DISTRIBUTION PLOTS (Quantitative Data)**

```python
sns.histplot(df['total_bill'], bins=40)
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, hue='sex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df,kind='reg')
```

```python
sns.pairplot(data=df, hue='day',palette='rainbow')
```

```python
sns.pairplot(data=df, hue='sex',palette='rainbow')
```

```python
sns.rugplot(data=df,x='total_bill', y='tip',hue='sex',palette='rainbow')
```
