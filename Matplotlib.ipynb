```python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^', t, t**4, 'yo')
plt.show()
```

```python
a = np.random.randint(1,51, 50)
b = np.random.randint(1,51, 50)
plt.scatter(a , b)
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [10, 20, 100]

plt.plot(names, values, 'ro')
plt.show()
```

```python
plt.bar(names, values, color='r')
plt.show()
```

```python
plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values, color='g')
plt.subplot(132)
plt.scatter(names, values, color='g')
plt.subplot(133)
plt.plot(names, values, color='g')
plt.suptitle('Categorical Plotting')
plt.show()
```
