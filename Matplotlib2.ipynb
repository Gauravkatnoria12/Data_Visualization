```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 5, 10)
y = x**2
```

```python
plt.plot(x, y)
plt.title("Hello")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

# **Subplots**

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
plt.subplot(2,2,1)
plt.plot(y, x, 'yo')
plt.subplot(2,2,2)
plt.plot(x,y,  'b-')
plt.subplot(2,2,3)
plt.plot(x,y**2, 'r-', marker="o")
plt.subplot(2,2,4)
plt.plot(x**2, y)
```

# **Object Oriented Way , Figures**

*Matlotlib Allows the aspect ratio, DPI and Figure size to be the specified when the figure abject is created. You can use the figsize and DPI keyword arguments.*

*figsize is type of tuple of width and height of the figure in inches, dpi is the dots-per-inch(pixel-per-inch).*

```python
fig = plt.figure(figsize=(10, 10), dpi=100)
axis1 = fig.add_axes([0.1, 0.1, 1, 1])
axis1.plot(x, y)

axis2 = fig.add_axes([0.2, 0.6, 0.4, 0.4])
axis2.plot(x**2, y)
```
