```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# **Plots used to visualize the distributon of quantitative data**

* displot/histplot
* joinplot
* pairplot
* rugplot

# **1. DistPlot**

```python
data = sns.load_dataset('tips')
data
```

```python
data['size'].unique()
```

```python
sns.displot(data['total_bill']) # default bins = 20
```

```python
sns.histplot(data['total_bill'], bins = 100)  # Same as Dist Plot
```

```python
sns.histplot(data['total_bill'], bins = 20, kde = True)
```

```python
plt.subplot(1, 2, 1)
sns.histplot(data['total_bill'], bins = 20, kde = True)
plt.subplot(1, 2, 2)
sns.histplot(data['tip'], kde=True)
```

```python
datas = sns.load_dataset('tips')
datas
```

# **2. JointPlot**

```python
sns.jointplot( x = 'total_bill', y = 'tip', data=datas, kind='scatter')
```

```python
sns.jointplot( x = 'total_bill', y = 'tip', data=datas, kind='kde')
```

```python
sns.jointplot( x = 'total_bill', y = 'tip', data=datas, kind='hex')
```

```python
sns.jointplot( x = 'total_bill', y = 'tip', data=datas, kind='reg')
```

# **3. PairPlot**

```python
sns.pairplot(datas)
```

```python
sns.pairplot(datas, hue='sex', palette='rainbow')
```

```python
sns.pairplot(datas, hue='size', palette='rainbow')
```

# **4. RugPlot**

```python
sns.rugplot(datas['total_bill'])
```

```python
sns.rugplot(datas)
```
