```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
df = pd.DataFrame(data)
df
```

```python
sns.barplot(df['size'].head())
```

```python
sns.barplot(data=df, x='sex', y='total_bill', estimator=np.sum)
```

```python
sns.lineplot(df['size'].head(50))
```

```python
sns.violinplot(x='day', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
```

```python
sns.stripplot(x='day', y='total_bill', data=df)
```

```python
sns.swarmplot(x='day', y='total_bill', data=df)
```

```python
core =df[['total_bill', 'tip', 'size']].corr()
sns.heatmap(core, annot=True) # annot for values in boxes
```

```python
sns.clustermap(core)
```

```python
sns.lmplot(x='tip', y='total_bill', data=df, hue='sex')
```
